```python
# convert data type of time to datetime
df[['time']] = df[['time']].apply(pd.to_datetime)
```

```python
#RAW WATER TURBIDITY TREND
plt.figure(figsize =(7, 4))
plt.plot(df[['time']], df[['turbidity']]) 
plt.legend(['TURBIDITY']);
plt.xticks(rotation=45);
#plt.ylim([0, 350]);
#plt.title('A graph of Turbidity against Time for the sensor-node at NYEWASCO-Raw Water Quality Section \n', size = 16)
plt.xlabel('Time (Date)', size = 14);
plt.ylabel('Turbidity (NTU)', size = 14);
#plt.tight_layout()
plt.savefig("TU_Dataset1.png", bbox_inches = 'tight', dpi=550)
```
