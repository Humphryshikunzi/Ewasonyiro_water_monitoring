```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame, Series
from pandas.io.json import json_normalize
from influxdb import InfluxDBClient
from datetime import datetime, timedelta, timezone
from tabulate import tabulate
```

```python
df = pd.read_csv (r'C:\Users\nahsh\desktop\wqm\JUPYTER\Hu\Data_Raw_Water.csv')
```

```python
# convert data type of time to datetime
df[['time']] = df[['time']].apply(pd.to_datetime)
```

```python
#RAW WATER TURBIDITY TREND
plt.figure(figsize =(20, 10))
plt.plot(df[['time']], df[['turbidity']]) 
plt.legend(['TURBIDITY']);
plt.xticks(rotation=45);
plt.ylim([0, 1100]);
plt.title('A graph of Turbidity against Time for the sensor-node at NYEWASCO-Raw Water Quality Section \n')
plt.xlabel('Time (Date)');
plt.ylabel('Turbidity (NTU)');
#plt.tight_layout()
```

```python
#RAW WATER pH TREND
plt.figure(figsize =(20, 10))
plt.plot(df[['time']], df[['pH']]) 
plt.rcParams['font.size'] = '20'
plt.legend(['pH Value']);
plt.xticks(rotation=45);
#plt.ylim([7.2, 7.6]);
plt.title('A graph of Water pH against Time for the sensor-node at NYEWASCO-Raw Water Quality Section \n')
plt.xlabel('Time (Date)');
plt.ylabel('Water pH');
#plt.tight_layout()
```
